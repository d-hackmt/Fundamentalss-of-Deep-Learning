```python
#python code for gradient descent 
# we have values of x and y vectors 
# we are going to derive the best fit line or an equation using m and b
# to know how well you are doing you should print cost at each iteration and you shpuld be reducing the cost ALWAYS 


import numpy as np
```

```python
# first thing  write a function , starting with some value of m and b and take steps one by one and define the number of iteration (how many baby steps)
def gradient_descent(x, y):
  m_curr = b_curr = 0  #we start by 0 
  iterations = 5000 
  n = len(x)
  learning_rate = 0.08  #this is trial and error its a parameter its a baby step which you have to decide and tune so yea 

  for i in range(iterations):
    y_pred = m_curr * x + b_curr 

    #lets print cost

    cost = 1/n * sum([val**2 for val in (y-y_pred)])  # also to deal with the negative values

    # m derivative and b derivative  (partial )

    md = -(2/n) * sum(x*(y-y_pred)) 
    bd = -(2/n) * sum(y-y_pred) 

    # now once we have our equation lets adjust our m and b current

    m_curr = m_curr - learning_rate * md 
    
    b_curr = b_curr - learning_rate * bd 

    #lets print the values at each iteration

    print ("m{} , b{} , cost {}, iteration {}".format(m_curr,b_curr,cost,i))


```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
